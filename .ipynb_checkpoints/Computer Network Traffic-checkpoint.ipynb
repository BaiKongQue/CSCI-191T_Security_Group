```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('~/Documents/CSCI/CS191T/csci-191T-security-group/cs448b_ipasn.csv')
```

```python
df.head(5)
```

```python
# Grouping logs of the same day
df['date']= pd.to_datetime(df['date'])
df = df.groupby(['date','l_ipn'],as_index=False).sum()
```

```python
# Obtain the day and week 
df['day'] = df['date'].dt.dayofyear
df['week'] = df['date'].dt.dayofweek
```

```python
ip0 = df[df['l_ipn']==0]
max0 = np.max(ip0['f'])
ip1 = df[df['l_ipn']==1]
max1 = np.max(ip1['f'])
ip2 = df[df['l_ipn']==2]
max2 = np.max(ip2['f'])
ip3 = df[df['l_ipn']==3]
max3 = np.max(ip3['f'])
ip4 = df[df['l_ipn']==4]
max4 = np.max(ip4['f'])
ip5 = df[df['l_ipn']==5]
max5 = np.max(ip5['f'])
ip6 = df[df['l_ipn']==6]
max6 = np.max(ip6['f'])
ip7 = df[df['l_ipn']==7]
max7 = np.max(ip7['f'])
ip8 = df[df['l_ipn']==8]
max8 = np.max(ip8['f'])
ip9 = df[df['l_ipn']==9]
max9 = np.max(ip9['f'])
```

```python
f,axarray = plt.subplots(5,2,figsize=(15,20))

axarray[0,0].plot(ip0['day'],ip0['f'])
axarray[0,0].plot(ip0['day'], [ip0['f'].mean() + 3*ip0['f'].std()]*len(ip0['day']),color='g')
axarray[0,0].set_title("Local IP 0 Flow")


axarray[0,1].plot(ip1['day'], ip1['f'])
axarray[0,1].plot(ip1['day'], [ip1['f'].mean() + 3*ip1['f'].std()]*len(ip1['day']),color='g')
axarray[0,1].set_title("Local IP 1 Flow")

axarray[1,0].plot(ip2['day'], ip2['f'])
axarray[1,0].set_title("Local IP 2 Flow")
axarray[1,0].plot(ip2['day'], [ip2['f'].mean() + 3*ip2['f'].std(ddof=0)]*len(ip2['day']),color='g')


axarray[1,1].plot(ip3['day'], ip3['f'])
axarray[1,1].set_title("Local IP 3 Flow")
axarray[1,1].plot(ip3['day'], [ip3['f'].mean() + 3*ip3['f'].std(ddof=0)]*len(ip3['day']),color='g')


axarray[2,0].plot(ip4['day'], ip4['f'])
axarray[2,0].set_title("Local IP 4 Flow")
axarray[2,0].plot(ip4['day'], [ip4['f'].mean() + 3*ip4['f'].std(ddof=0)]*len(ip4['day']),color='g')

axarray[2,1].plot(ip5['day'], ip5['f'])
axarray[2,1].set_title("Local IP 5 Flow")
axarray[2,1].plot(ip5['day'], [ip5['f'].mean() + 3*ip5['f'].std(ddof=0)]*len(ip5['day']),color='g')

axarray[3,0].plot(ip6['day'], ip6['f'])
axarray[3,0].set_title("Local IP 6 Flow")
axarray[3,0].plot(ip6['day'], [ip6['f'].mean() + 3*ip6['f'].std(ddof=0)]*len(ip6['day']),color='g')

axarray[3,1].plot(ip7['day'], ip7['f'])
axarray[3,1].set_title("Local IP 7 Flow")
axarray[3,1].plot(ip7['day'], [ip7['f'].mean() + 3*ip7['f'].std(ddof=0)]*len(ip7['day']),color='g')

axarray[4,0].plot(ip8['day'], ip8['f'])
axarray[4,0].set_title("Local IP 8 Flow")
axarray[4,0].plot(ip8['day'], [ip8['f'].mean() + 3*ip8['f'].std(ddof=0)]*len(ip8['day']),color='g')


axarray[4,1].plot(ip9['day'], ip9['f'])
axarray[4,1].set_title("Local IP 9 Flow")
axarray[4,1].plot(ip9['day'], [ip9['f'].mean() + 3*ip9['f'].std(ddof=0)]*len(ip9['day']),color='g')
```
